```python
import pandas as pd
```

```python
df = pd.read_excel('/Users/a19028558/Desktop/Ralf_Hack/furniture_2020.xls')
```

```python
df.columns = df.iloc[1]
df = df[2:]
df.head()
```

```python
df_transponse = df.T.reset_index()
```

```python
df_raif = pd.read_csv('/Users/a19028558/Desktop/Ralf_Hack/train.csv')
# df_raif['date'].value_counts()
```

```python
dates_raif = set(df_raif['date'])
```

```python
df_transponse.columns = df_transponse.iloc[3]
df_transponse = df_transponse[4:]
```

```python
df_transponse.rename(columns={df_transponse.columns[0]:'date'}, inplace=True)
```

```python

date_collector = {}
for i, date in enumerate(df_transponse['date']):
    i += 1
    key = f'0{i}' if i < 10 else str(i)
    date_collector[key] = date
    
date_collector
```

```python
mapping = {}
for date in list(dates_raif):
    if date.split('-')[1] in date_collector:
        mapping[date] = date_collector[date.split('-')[1]]
    else:
        print('f')
```

```python
df_raif['month'] = df_raif['date'].apply(lambda x: mapping[x])
```

```python
df_transponse.rename(columns={'date':'month'}, inplace=True)
```

```python
df_result = pd.merge(df_raif, df_transponse, on='month', how='left')
df_result.drop(columns='month').to_csv('additional_metrics.csv', index=False)
```
